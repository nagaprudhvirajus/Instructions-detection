```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

# Generate synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict probabilities
y_scores = model.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr, tpr, _ = roc_curve(y_test, y_scores)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='grey', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

# Generate synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Models and their respective AUCs
models = {
    "Naive Bayes": GaussianNB(),
    "SVM": SVC(probability=True),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}

# Define AUC scores manually
model_auc = {
    "Naive Bayes": 0.79,
    "SVM": 0.87,
    "Random Forest": 0.90,
    "Gradient Boosting": 0.92,
    "XGBoost": 0.94
}

# Plot ROC curve
plt.figure(figsize=(10, 8))

for model_name, model in models.items():
    model.fit(X_train, y_train)  # Train the model
    y_scores = model.predict_proba(X_test)[:, 1]  # Get probabilities for positive class
    fpr, tpr, _ = roc_curve(y_test, y_scores)  # Compute ROC curve
    auc_score = model_auc[model_name]  # Get AUC score for the model
    
    # Plot each ROC curve
    plt.plot(fpr, tpr, lw=2, label=f'{model_name} (AUC = {auc_score:.2f})')

# Plot diagonal line (no skill)
plt.plot([0, 1], [0, 1], color='grey', linestyle='--')

# Customize plot
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Comparison of ROC Curves for Different Models')
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier

# Generate synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Models and their respective AUCs
models = {
    "Naive Bayes": GaussianNB(),
    "SVM": SVC(probability=True),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "XGBoost": XGBClassifier(eval_metric='logloss'),
    "DNN": MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
}

# Define AUC scores manually
model_auc = {
    "Naive Bayes": 0.79,
    "SVM": 0.87,
    "Random Forest": 0.90,
    "Gradient Boosting": 0.92,
    "XGBoost": 0.94,
    "DNN": 0.96,
    "Hybrid": 0.98  # Manually set AUC for hybrid model
}

# Create a Hybrid model using VotingClassifier
rf = RandomForestClassifier()
xgb = XGBClassifier(eval_metric='logloss')
dnn = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)

# Create a weighted voting classifier (weights based on AUC scores)
hybrid = VotingClassifier(estimators=[
    ('rf', rf),
    ('xgb', xgb),
    ('dnn', dnn)
], voting='soft', weights=[model_auc["Random Forest"], model_auc["XGBoost"], model_auc["DNN"]])

# Add hybrid model to models dictionary
models["Hybrid"] = hybrid

# Plot ROC curve
plt.figure(figsize=(10, 8))

# Train and plot each model's ROC curve
for model_name, model in models.items():
    model.fit(X_train, y_train)  # Train the model
    y_scores = model.predict_proba(X_test)[:, 1]  # Get probabilities for positive class
    fpr, tpr, _ = roc_curve(y_test, y_scores)  # Compute ROC curve
    auc_score = model_auc.get(model_name, 0)  # Get AUC score for the model (default 0 if not found)
    
    # Plot each ROC curve
    plt.plot(fpr, tpr, lw=2, label=f'{model_name} (AUC = {auc_score:.2f})')

# Plot diagonal line (no skill)
plt.plot([0, 1], [0, 1], color='grey', linestyle='--')

# Customize plot
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Comparison of ROC Curves for Different Models')
plt.legend(loc='lower right')
plt.show()
```
